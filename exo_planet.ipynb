```python
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
search_result=lk.search_lightcurve('Kepler-10', mission='Kepler')
search_result
```

```python
lc=search_result[0].download()
lc.plot(title="Kepler-10 Light Curve")
plt.show()
```

```python
lc_clean=lc.remove_nans().flatten(window_length=401)
lc_clean.plot(title="Flattened Light Curve (Noise Reduced)")
plt.show()
```

```python
import numpy as np
periodogram=lc_clean.to_periodogram(method='bls', period=np.linspace(0.5,20,10000))
periodogram.plot(title="Transit Periodogram")
plt.show()
```

```python
best_period=periodogram.period_at_max_power
print(f"Best period (planet's orbital period): {best_period}")
folded_lc=lc_clean.fold(period=best_period)
folded_lc.plot(title=f"Phase Folded Light Curve (Period={best_period.value:.2f} days)")
plt.show()
```

```python
depth=1-folded_lc.flux.min()
print(f"Transit depth: {depth:.6f}")
```

```python
Rp_Rs=np.sqrt(depth)
print(f"Planet-to-star ratio (Rp/Rs): {Rp_Rs:.4f}")
```

```python
R_star=1.0
R_planet=Rp_Rs*R_star
print(f"Approx planet radius: {R_planet:.3f} RO (~{R_planet*109:.1f} R_earth)")
```

```python

G = 6.67430e-11
M_sun = 1.989e30
R_sun = 6.957e8

M_star = 0.97 * M_sun
P = best_period.value * 24 * 3600

a = ((G * M_star * (P**2)) / (4 * np.pi**2))**(1/3)
a_AU = a / (1.496e11)

print(f"Semi-major axis ≈ {a_AU:.2f} AU")
```

Using NASA Kepler telescope data (Kepler-10 star),
I detected a periodic brightness dip every ~0.84 days.
This indicates the presence of an exoplanet, Kepler-10b,
orbiting extremely close to its star (~0.02 AU).
